## Now training ML model to predict the future sales of a product but before training let's split the data into training and test sets.

```python
x = np.array(data.drop(["Sales"],1))
y = np.array(data['Sales'])
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2,random_state=42)
```

## Now train the model to predict future sales

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
print(model.score(xtest, ytest))
```

```python
# future_sales = [[TV, Social, Newspaper]]
TV = int(input('Enter Money Spent On Marketing over TeleVision'))
SM = int(input('Enter Money Spent On Marketing over Social Media'))
NEP = int(input('Enter Money Spent On Marketing over Newspaper'))
future_sales = np.array([[TV, SM, NEP]])
print('Number of Units Sold ',model.predict(future_sales))
```
